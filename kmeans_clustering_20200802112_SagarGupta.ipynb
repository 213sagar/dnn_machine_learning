```python
## importing required library 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
## Start by  visualizing the data points 
x=[4,5,22,43,54,5,53,3,33,34]
y=[24,2,34,23,10,34,19,34,22,23]

plt.scatter(x,y)
plt.show()
```

```python
## now we are finding the k value from the Elbow method 
from sklearn.cluster import KMeans
data=list(zip(x,y))
inertias=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
    

plt.plot(range(1,11),inertias,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number Of Clusters ')
plt.ylabel("INertia")
plt.show()
```

### Here we see in the graph from the point 2 onwards the variance is decreasing in the same pattern. So our K value is 2.

```python
## after the k=2 , Visualiztion of the clusters 
kmeans=KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x,y,c=kmeans.labels_)## kmeans.label - different types of color according to the cluster values of K
plt.show()
```
